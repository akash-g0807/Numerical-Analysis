```python
# Setup function
def f(x):
    return np.sin(x) * (1 - x)

# Exact derivative value at x = 1
exact = -np.sin(1)

# Forward difference function
def delta_f(h):
    return (f(1 + h) - f(1)) / h


# Create table header
hdr1 = "h"
hdr2 = "Δf(1)"
hdr3 = "|f'(1) − Δf(1)|"
hdr4 = "Error reduction factor"

print(f"{hdr1:<20} | {hdr2:<18} | {hdr3:<18} | {hdr4:<18}")
print("-" * 85)

error_prev = None

for n in range(1, 11):

    h = 2**(-n)
    h_str = f"2^-{n} = {h:g}"

    approx = delta_f(h)
    error = abs(exact - approx)

    if error_prev is None:
        print(f"{h_str:<20} | {approx:<18.15f} | {error:<18.15f} | ")
    else:
        reduction = error_prev / error
        print(f"{h_str:<20} | {approx:<18.15f} | {error:<18.15f} | {reduction:<18.15f}")

    error_prev = error
```

```python
def plot_forward_difference_errors(f, x, exact, H):

    # Create list of errors
    AbsError = [] # start off with a blank list of errors
    for h in H:
        approx = (f(x+h) - f(x))/h
        AbsError.append(abs(approx - exact))

    # Make a loglog plot
    plt.loglog(H, AbsError, 'b*')
    plt.xlabel("h")
    plt.ylabel("Absolute Error")
    plt.title("Absolute Error vs. h")
    plt.grid()
    return(plt)
```

```python
def ForwardDiff(f, a, b, N):
    # Create N+1 equally spaced partition points from a to b.
    # We need N+1 points because N subintervals have one more endpoint than intervals.
    x = np.linspace(a, b, N+1)

    # Compute the step size h using the difference between consecutive partition points.
    h = x[1] - x[0]

    # Create an empty list that will store the forward-difference approximations.
    df = []

    # Loop through all points except the last one, since forward difference
    # needs the next point x[j+1].
    for j in np.arange(len(x)-1):

        # Append the forward-difference approximation
        # (f(x[j+1]) - f(x[j])) / h to the list.
        df.append((f(x[j+1]) - f(x[j])) / h)

    # Return the list of derivative approximations.
    return df
```

```python
f = lambda x: np.sin(x)
exact_df = lambda x: np.cos(x)

a = 0
b = 2*np.pi
N = 100  # number of subintervals

x = np.linspace(a,b,N+1)
# Creates N+1 equally spaced points between a and b.
# We need N+1 points because N subintervals require N+1 endpoints.

df = ForwardDiff(f,a,b,N)
# Calls the ForwardDiff function to compute the forward-difference
# approximation of the derivative at each point except the last one.

# Plot the function, the exact derivative, and the approximation
plt.plot(x, f(x), 'b', x, exact_df(x), 'r--', x[0:-1], df, 'k-.')
plt.grid()

plt.legend(['f(x) = sin(x)',
            'exact first deriv',
            'approx first deriv'])

plt.show()
```

```python
def ForwardDiff2(f, a, b, N):
    x = np.linspace(a, b, N+1)
    h = x[1] - x[0]
    y = f(x)
    df = (y[1:] - y[:-1]) / h
    return df
```

```python
def plot_max_forward_difference_errors(f, f_prime, a, b, H):

    max_errors = []

    for h in H:

        N = int((b-a)/h)

        x = np.linspace(a,b,N+1)

        df = ForwardDiff(f,a,b,N)

        error = abs(f_prime(x[0:-1]) - df)

        max_errors.append(np.max(error))

    plt.loglog(H, max_errors, 'b*-')
    plt.xlabel("h")
    plt.ylabel("Max Error")
    plt.grid()
    plt.title("Maximum Forward Difference Error")
    plt.show()
    slope, intercept = np.polyfit(np.log(H), np.log(max_errors),1)
    print(slope)
```

```python
import numpy as np

# Setup function
def f(x):
    return np.exp(np.sin(x**(3/2)))

# Exact derivative value at x = 1
exact = -np.sin(1)

# Central difference function
def delta_f(h):
    return (f(1 + h) -2*f(1) + f(1 - h)) / (h**2)


# Create table header
hdr1 = "h"
hdr2 = "δf(1)"
hdr3 = "|f'(1) − δf(1)|"
hdr4 = "Error reduction factor"

print(f"{hdr1:<20} | {hdr2:<18} | {hdr3:<18} | {hdr4:<18}")
print("-" * 85)

error_prev = None

for n in range(1, 11):

    h = 0.1
    h_str = f"2^-{n} = {h:g}"

    approx = delta_f(h)
    error = abs(exact - approx)

    if error_prev is None:
        print(f"{h_str:<20} | {approx:<18.15f} | {error:<18.15f} | ")
    else:
        reduction = error_prev / error
        print(f"{h_str:<20} | {approx:<18.15f} | {error:<18.15f} | {reduction:<18.15f}")

    error_prev = error
```

```python
def plot_central_difference_errors(f, x, exact, H):

    # Create list of errors
    AbsError = []  # start off with a blank list of errors
    for h in H:
        approx = (f(x+h) - f(x-h)) / (2*h)
        AbsError.append(abs(approx - exact))

    # Make a loglog plot
    plt.loglog(H, AbsError, 'b*')
    plt.xlabel("h")
    plt.ylabel("Absolute Error")
    plt.title("Central Difference Absolute Error vs. h")
    plt.grid()
    return(plt)

H = [2**(-n) for n in range(1,11)]

plot_central_difference_errors(f, 1, exact, H)
plt.show()
```

```python
def CentralDiff(f, a, b, N):

    x = np.linspace(a, b, N+1)   # grid points
    h = x[1] - x[0]              # step size

    y = f(x)                     # evaluate function

    df = (y[2:] - y[:-2]) / (2*h)   # central difference formula

    return df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f
def f(x):
    return np.sin(x)

# Define its derivative
def df(x):
    return np.cos(x)

# Choose interval
a = 0
b = 2*np.pi

m = 16  # number of different step sizes

# Pre-allocate vectors for errors
fd_error = np.zeros(m)
cd_error = np.zeros(m)

# Pre-allocate vector for step sizes
H = np.zeros(m)

# Loop over step sizes
for n in range(m):

    N = 2**(n+2)  # number of subintervals

    x = np.linspace(a, b, N+1)

    y = f(x)

    h = x[1] - x[0]  # step size

    # Forward difference
    forward_diff = (y[1:] - y[:-1]) / h

    # Central difference
    central_diff = (y[2:] - y[:-2]) / (2*h)

    # Compute maximum errors
    fd_error[n] = max(abs(forward_diff - df(x[:-1])))
    cd_error[n] = max(abs(central_diff - df(x[1:-1])))

    # Store step size
    H[n] = h


# Plot results
plt.loglog(H, fd_error, 'b-', label='Approximation Method A')
plt.loglog(H, cd_error, 'r-', label='Approximation Method B')

plt.xlabel('Step size h')
plt.ylabel('Maximum Absolute Error')
plt.title('Comparing Two First Derivative Approximations')

plt.grid()
plt.legend()
plt.show()
```

```python
def SecondDiff(f, a, b, N):
    x = np.linspace(a, b, N+1)
    h = x[1] - x[0]
    y = f(x)
    ddf = (y[2:] - 2*y[1:-1] + y[:-2]) / h**2
    return ddf
```

# Integration

```python
def RiemannSum(f, a, b, N, method='left'):
    x = np.linspace(a, b, N + 1)
    dx = x[1] - x[0]

    if method == 'left':
        return np.sum(f(x[:-1])) * dx

    elif method == 'right':
        return np.sum(f(x[1:])) * dx

    elif method == 'midpoint':
        midpoints = (x[:-1] + x[1:]) / 2
        return np.sum(f(midpoints)) * dx

    else:
        raise ValueError("method must be 'left', 'right', or 'midpoint'")
```

```python
def f1(x):
    return x**2

def f2(x):
    return np.sin(x)

print(RiemannSum(f1, 0, 1, 100, method='left'))      # approx 1/3
print(RiemannSum(f1, 0, 1, 100, method='right'))     # approx 1/3
print(RiemannSum(f1, 0, 1, 100, method='midpoint'))  # approx 1/3

print(RiemannSum(f2, 1, 6, 25, method='midpoint'))  # approx 2
```

```python
import numpy as np

# Function
def f(x):
    return np.sin(x)

# Exact integral
exact = 1 - np.cos(1)

# Table header
hdr1 = "Δx"
hdr2 = "A_left(Δx)"
hdr3 = "|I - A_left(Δx)|"
hdr4 = "Error reduction factor"

print(f"{hdr1:<20} | {hdr2:<18} | {hdr3:<18} | {hdr4:<18}")
print("-" * 82)

error_prev = None

for n in range(1, 11):
    dx = 3**(-n)
    dx_str = f"3^-{n} = {dx:g}"

    N = int(round(1 / dx))
    approx = RiemannSum(f, 0, 1, N, method='left')
    error = abs(exact - approx)

    if error_prev is None:
        print(f"{dx_str:<20} | {approx:<18.15f} | {error:<18.15f} | ")
    else:
        reduction = error_prev / error
        print(f"{dx_str:<20} | {approx:<18.15f} | {error:<18.15f} | {reduction:<18.15f}")

    error_prev = error
```

```python
def plot_riemann_errors(f, a, b, exact, DX):
    # Create lists of errors
    error_left = []
    error_right = []
    error_midpoint = []

    for dx in DX:
        N = int(np.round((b - a) / dx))
        error_left.append(abs(RiemannSum(f, a, b, N, method='left') - exact))
        error_right.append(abs(RiemannSum(f, a, b, N, method='right') - exact))
        error_midpoint.append(abs(RiemannSum(f, a, b, N, method='midpoint') - exact))

    # Make a loglog plot
    plt.loglog(DX, error_left, 'b*', label='Left Sum', markersize=14)
    plt.loglog(DX, error_right, 'ro', label='Right Sum')
    plt.loglog(DX, error_midpoint, 'g+', label='Midpoint Sum')

    plt.xlabel("Δx")
    plt.ylabel("Absolute Error")
    plt.title("Absolute Error vs. Δx")
    plt.legend()
    plt.grid(True)
    return plt
```

```python
# Setup function and exact integral again
f = lambda x: np.sin(x)
exact = -np.cos(1) + np.cos(0)

# Create vector of step sizes
DX = [2**(-n) for n in range(1, 11)]

# Plot the absolute error
plot_riemann_errors(f, 0, 1, exact, DX)
```

```python
def Trapezoidal(f, a, b, N):
    x = np.linspace(a, b, N + 1)
    h = x[1] - x[0]
    y = f(x)
    return (h / 2) * (y[0] + 2*np.sum(y[1:-1]) + y[-1])
```

```python
def f1(x):
    return x**2

def f2(x):
    return np.sin(x)

print(Trapezoidal(f1, 0, 1, 100))      # should be close to 1/3
print(Trapezoidal(f2, 1, 4, 100))  # should be close to 2
```

```python
def Simpsons(f, a, b, N):
    x = np.linspace(a, b, N + 1)
    h = x[1] - x[0]

    left_right = 0.5 * (f(x[0]) + f(x[-1]))
    interior = np.sum(f(x[1:-1]))
    midpoints = np.sum(f((x[:-1] + x[1:]) / 2))

    return (h / 3) * (left_right + interior + 2 * midpoints)
```

```python
def f1(x):
    return 1

def f2(x):
    return x

def f3(x):
    return x**2

def f4(x):
    return np.cos(x)

print(Simpsons(f1, 0, 1, 10))         # exact = 1
print(Simpsons(f2, 0, 1, 10))         # exact = 1/2
print(Simpsons(f3, 0, 1, 10))         # exact = 1/3
print(Simpsons(f4, 2, 3, 10))     # exact = 2
```
